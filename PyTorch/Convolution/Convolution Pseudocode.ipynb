```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot
jtplot.style(theme="monokai", context="notebook", ticks=True)
np.random.seed(42)
```

```python
X = np.random.randint(1, 10, size=(4,4))
kernel = np.random.randint(1, 10, size=(2,2))
print("Input data")
print(X)

print("-"*10)
print("Kernel")
print(kernel)
```

```python
def convolve(X, kernel):
    
    """
    ------------------------------------------
    Description : 
    Computes the convolution between an image and it's respective kernel
    
    Parameters :
    X -- a numpy nd-array representing the image (n x n) size
    kernel -- a numpy nd-array representing the filter (m x m) size
    
    
    Return:
    out -- a numpy nd-array representing the convolved output (n - m + 1 x n - m + 1) size
    -------------------------------------------
    """
    
    input_size = X.shape
    kernel_size = kernel.shape
    
    out_shape = (input_size[0] - kernel_size[0] + 1, input_size[1] - kernel_size[1] + 1)
    out = np.zeros(out_shape)
    
    for i in range(out.shape[0]):
        for j in range(out.shape[1]):
            out[i,j] = np.sum(np.multiply(X[i:i+kernel_size[0], j:j+kernel_size[0]], kernel))
    
    return out
    
    
```

```python
output = convolve(X, kernel)
print("Convolution Output")
print(output)
```
